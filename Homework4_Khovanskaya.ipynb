# 3 задача

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
data = houses.groupby('floors').agg({'price':'mean'})
data.reset_index()
plt.bar(data.index, data['price'])
plt.title('Средняя стоимость в зависимости от этажности дома')
plt.xlabel('Количество этажей')
plt.ylabel('Средняя стоимость домов');
plt.grid()
```

```python
data = houses.groupby('waterfront').agg({'price':'mean'})
data.reset_index()
plt.figure(figsize=(3,3))
plt.bar(data.index, data['price'])
plt.title('Средняя стоимость в зависимости от наличия вида на набережную')
plt.xlabel('Есть ли вид на набережную')
plt.ylabel('Средняя стоимость домов');
```

```python
data = houses.groupby('view').agg({'price':'mean'})
data.reset_index()
plt.bar(data.index, data['price'])
plt.title('Средняя стоимость в зависимости от вида')
plt.xlabel('Оценка вида')
plt.ylabel('Средняя стоимость домов');
```

```python
data = houses.groupby('condition').agg({'price':'mean'})
data.reset_index()
plt.bar(data.index, data['price'])
plt.title('Средняя стоимость в зависимости от оценки состояния')
plt.xlabel('Оценка состояния')
plt.ylabel('Средняя стоимость домов');
```

```python
data = houses.groupby('grade').agg({'price':'mean'})
data.reset_index()
plt.bar(data.index, data['price'])
plt.title('Средняя стоимость в зависимости от класса grade')
plt.xlabel('grade')
plt.ylabel('Средняя стоимость домов');
```

```python
data = houses.groupby('yr_built').agg({'price':'mean'})
data.reset_index()
plt.figure(figsize=(20,9))
plt.bar(data.index, data['price'])
plt.title('Средняя стоимость в зависимости от года постойки')
plt.xlabel('Год постойки')
plt.ylabel('Средняя стоимость домов');
```

**По полученным графикам можно сделать вывод:**
- стоимость недвижимости зависит от оценки вида, чем выше оценка, тем выше стоимость В том числе наличие вида на набережную значительно увеличивает стоимость;
- также на стоимость влияет оценка состояния недвижимости, чем выше оценка, тем выше стоимость;
- стоимость объектов недвижимости выше, если они относятся к более высокому классу (grade);
- прослеживается зависимость стоимости от года постойки: средняя стоимость исторических объектов с годом постойки до 1920 года и новых объектов. постоенных после 2000 года выше средней стоимости объектов, постоенный в период с 1920 по 2000 годы;
- этажность дома также оказывает влияние на стоимостьЖ средняя стоимость домов с этажностью 2,5 и 3,5 выше, чем в остальных домах
